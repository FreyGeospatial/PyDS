```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
stats = pd.read_csv("DemographicData.csv")
stats
```

```python
#linear model
sns.lmplot(data = stats, x = 'Internet users', y = "Birth rate", fit_reg=False) #fit_reg removes line
```

```python
sns.boxplot(data = stats, x = "Income Group", y = "Birth rate")
```

```python
sns.distplot(stats["Internet users"], bins=30)
```

```python
sns.distplot(stats["Birth rate"], bins=30)
```

```python
sns.lmplot(data = stats, x = 'Internet users', y = "Birth rate", fit_reg=False, \
          hue = 'Income Group') #hue is like ggplot color within aes
```

```python
sns.lmplot(data = stats, x = 'Internet users', y = "Birth rate", fit_reg=False, \
          hue = 'Income Group', col = 'Income Group') #hue is like ggplot color within aes
```

```python
##Keyword arguments
```

```python
sns.lmplot(data = stats, x = 'Internet users', y = "Birth rate", fit_reg=False, \
          hue = 'Income Group', height=7, \
          scatter_kws = {"s":100}) #keyword argument
```
